```python
from IPython.display import display, clear_output
import ipywidgets as widgets
from ipywidgets import Layout, HBox, VBox

out1 = widgets.Output()
out2 = widgets.Output()
out3 = widgets.Output()
```

```python
def box_layout():
    box_layout = Layout(
        display="flex",
        flex_flow="row wrap",
        align_items="center",
        justify_content="center",
        border="solid",
        width="100%",
    )

    vbox_layout = Layout(
        display="flex",
        flex_flow="column",
        align_items="center",
        align_self="center",
    )
    Vbox = VBox(children=[out2, out3], layout=vbox_layout)
    box = HBox(children=[Vbox, out1], layout=box_layout)
    return box
```

```python
def widgets_basic_filters(filter_function):
    # criação de caixas para deixar mais fácil a visualização das respostas
    box = box_layout()

    def submit(change):
        clear_output()
        out1.clear_output()
        out2.clear_output()
        out3.clear_output()
        display(filter_chose_class, file_name, order)
        if (
            filter_chose_class.value == "bandpass"
            or filter_chose_class.value == "bandstop"
        ):
            display(cutoff_frequency_low, cutoff_frequency_high, submit_button)
            run_filter(
                filter_chose_class.value,
                int(cutoff_frequency_low.value),
                file_name.value,
                int(order.value),
                filter_function,
                int(cutoff_frequency_high.value),
            )
        else:
            display(cutoff_frequency, submit_button)
            run_filter(
                filter_chose_class.value,
                int(cutoff_frequency.value),
                file_name.value,
                int(order.value),
                filter_function,
            )
        display(box)

    # criação de widgets para alterar de forma interativa os filtros
    file_name = widgets.Text(
        value="symphony.wav", description="File name", continuous_update=False
    )
    filter_chose_class = widgets.Dropdown(
        options=["lowpass", "highpass", "bandpass", "bandstop"],
        description="Filter class:",
        disabled=False,
    )
    cutoff_frequency = widgets.Text(
        value="3000", description="Cutoff frequency", continuous_update=False
    )
    cutoff_frequency_low = widgets.Text(
        value="2000", description="Low Cutoff frequency", continuous_update=False
    )
    cutoff_frequency_high = widgets.Text(
        value="5000", description="High Cutoff frequency", continuous_update=False
    )
    order = widgets.Text(value="5", description="Order", continuous_update=False)

    submit_button = widgets.Button(description="Atualizar", button_style="success")
    submit_button.on_click(submit)

    display(filter_chose_class, file_name, order, cutoff_frequency, submit_button)
    display(box)
```

```python
def widgets_equalizers():
    # criação de caixas para deixar mais fácil a visualização das respostas
    box = box_layout()

    def submit(change):
        clear_output()
        out1.clear_output()
        out2.clear_output()
        out3.clear_output()
        display(filter_chose_class, filter_chose_type, file_name, gain, submit_button)

        if filter_chose_type.value == "peak":
            display(center_frequency, bandwidth, submit_button)
            run_filter(
                0,
                file_name.value,
                filter_chose_type.value,
                int(gain.value),
                filter_chose_class.value,
                int(center_frequency.value),
                int(bandwidth.value),
            )
        else:
            display(cutoff_frequency, submit_button)
            run_filter(
                int(cutoff_frequency.value),
                file_name.value,
                filter_chose_type.value,
                int(gain.value),
                filter_chose_class.value,
            )
        display(box)

    # criação de widgets para alterar de forma interativa os filtros
    filter_chose_class = widgets.Dropdown(
        options=["lowpass", "highpass"],
        description="Filter class:",
        disabled=False,
    )
    filter_chose_type = widgets.Dropdown(
        options=["shelving", "peak"],
        description="Filter type:",
        disabled=False,
    )
    file_name = widgets.Text(
        value="symphony.wav", description="File name", continuous_update=False
    )
    cutoff_frequency = widgets.Text(
        value="5000", description="Cutoff frequency", continuous_update=False
    )
    gain = widgets.Text(value="6", description="Gain", continuous_update=False)
    center_frequency = widgets.Text(
        value="11000", description="Center frequency", continuous_update=False
    )
    bandwidth = widgets.Text(
        value="5000", description="Bandwidth", continuous_update=False
    )

    submit_button = widgets.Button(description="Atualizar", button_style="success")
    submit_button.on_click(submit)

    display(
        filter_chose_class,
        filter_chose_type,
        file_name,
        gain,
        cutoff_frequency,
        submit_button,
    )
    display(box)
```

```python
def widgets_wah():
    # criação de caixas para deixar mais fácil a visualização das respostas
    box = box_layout()

    def submit(change):
        clear_output()
        out1.clear_output()
        out2.clear_output()
        out3.clear_output()
        display(file_name, min_frequency, max_frequency, damp, wah_frequency, submit_button)
        wah_filter(
            float(damp.value),
            int(min_frequency.value),
            int(max_frequency.value),
            file_name.value,
            int(wah_frequency.value),
        )
        display(box)

    # criação de widgets para alterar de forma interativa os filtros

    file_name = widgets.Text(
        value="piano.wav", description="File name", continuous_update=False
    )
    min_frequency = widgets.Text(
        value="300", description="Cutoff frequency", continuous_update=False
    )
    max_frequency = widgets.Text(
        value="5000", description="Cutoff frequency", continuous_update=False
    )
    damp = widgets.Text(value="0.05", description="Damping", continuous_update=False)
    wah_frequency = widgets.Text(
        value="2000", description="Center frequency", continuous_update=False
    )

    submit_button = widgets.Button(description="Atualizar", button_style="success")
    submit_button.on_click(submit)

    display(file_name, min_frequency, max_frequency, damp, wah_frequency, submit_button)
    display(box)
```
