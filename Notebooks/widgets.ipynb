```python
from IPython.display import display, clear_output
import ipywidgets as widgets
from ipywidgets import Layout, HBox, VBox

out1 = widgets.Output()
out2 = widgets.Output()
out3 = widgets.Output()

style = {"description_width": "initial"}

with open("../Assets/loading-gif.gif", "rb") as file:
    # read file as string into `image`
    image = file.read()
```

```python
def box_layout():
    box_layout = Layout(
        display="flex",
        flex_flow="row wrap",
        align_items="center",
        justify_content="center",
        border="solid",
        width="100%",
    )

    vbox_layout = Layout(
        display="flex",
        flex_flow="column",
        align_items="center",
        align_self="center",
    )
    Vbox = VBox(children=[out2, out3], layout=vbox_layout)
    box = HBox(children=[Vbox, out1], layout=box_layout)
    return box
```

```python
def widgets_basic_filters():
    # criação de caixas para deixar mais fácil a visualização das respostas
    box = box_layout()

    def basic_filters_layout():
        clear_output()
        out1.clear_output()
        out2.clear_output()
        out3.clear_output()
        with loading_widget:
            display(loading_gif)
        display(filter_chose_class, file_name, order)
        if (
            filter_chose_class.value == "bandpass"
            or filter_chose_class.value == "bandstop"
        ):
            display(cutoff_frequency_range, submit_button)
            display(loading_widget)
            run_filter(
                filter_chose_class.value,
                cutoff_frequency_range.value[0],
                file_name.value,
                order.value,
                cutoff_frequency_range.value[1],
            )
        else:
            display(cutoff_frequency, submit_button)
            display(loading_widget)
            run_filter(
                filter_chose_class.value,
                cutoff_frequency.value,
                file_name.value,
                order.value
            )
        loading_widget.clear_output()
        display(box)

    def change(change):
        if change["type"] == "change" and change["name"] == "value":
            basic_filters_layout()

    def submit(change):
        basic_filters_layout()

    # criação de widgets para alterar de forma interativa os filtros
    file_name = widgets.Text(
        value="symphony.wav",
        description="Nome do arquivo:",
        continuous_update=False,
        style=style,
    )
    filter_chose_class = widgets.Dropdown(
        options=["highpass", "lowpass", "bandpass", "bandstop"],
        description="Classe do filtro:",
        disabled=False,
        style=style,
    )
    cutoff_frequency = widgets.IntSlider(
        value=3000,
        min=500,
        max=22000,
        step=500,
        description="Freq. de corte:",
        disabled=False,
        continuous_update=False,
        orientation="horizontal",
        readout=True,
        readout_format="d",
        style=style,
    )
    cutoff_frequency_range = widgets.IntRangeSlider(
        value=[3000, 5000],
        min=500,
        max=22000,
        step=500,
        description="Freq. de corte:",
        disabled=False,
        continuous_update=False,
        orientation="horizontal",
        readout=True,
        readout_format="d",
        style=style,
    )

    order = widgets.BoundedIntText(
        value=5,
        min=1,
        max=10,
        description="Ordem",
        continuous_update=False,
        style=style,
    )

    loading_widget = widgets.Output()
    loading_gif = widgets.Image(
        value=image,
        format="gif",
        width=100,
        height=100,
    )

    submit_button = widgets.Button(description="Atualizar", button_style="success")
    submit_button.on_click(submit)
    filter_chose_class.observe(change)
    display(filter_chose_class, file_name, order, cutoff_frequency, submit_button)
    display(box)
```

```python
def widgets_equalizers():
    # criação de caixas para deixar mais fácil a visualização das respostas
    box = box_layout()

    def change(change):
        if change["type"] == "change" and change["name"] == "value":
            if filter_chose_type.value == "peak":
                filter_chose_class.disabled = True
            else:
                filter_chose_class.disabled = False

    def submit(change):
        clear_output()
        out1.clear_output()
        out2.clear_output()
        out3.clear_output()
       
        with loading_widget:
            display(loading_gif)
        if filter_chose_type.value == "peak":
            display(filter_chose_class, filter_chose_type, file_name, gain, center_frequency, bandwidth, submit_button)
            display(loading_widget)
            run_filter(
                0,
                file_name.value,
                filter_chose_type.value,
                gain.value,
                filter_chose_class.value,
                center_frequency.value,
                bandwidth.value,
            )
        else:
            display(filter_chose_class, filter_chose_type, file_name, gain, cutoff_frequency, submit_button)
            display(loading_widget)
            run_filter(
                cutoff_frequency.value,
                file_name.value,
                filter_chose_type.value,
                gain.value,
                filter_chose_class.value,
            )
        loading_widget.clear_output()
        display(box)

    # criação de widgets para alterar de forma interativa os filtros
    filter_chose_class = widgets.Dropdown(
        options=["lowpass", "highpass"],
        description="Classe do filtro:",
        disabled=False,
        style=style,
    )
    filter_chose_type = widgets.Dropdown(
        options=["shelving", "peak"],
        description="Tipo do filtro:",
        disabled=False,
        style=style,
    )
    file_name = widgets.Text(
        value="piano.wav",
        description="Nome do arquivo:",
        continuous_update=False,
        style=style,
    )
    cutoff_frequency = widgets.IntSlider(
        value=3000,
        min=500,
        max=22000,
        step=500,
        description="Freq. de corte:",
        disabled=False,
        continuous_update=False,
        orientation="horizontal",
        readout=True,
        readout_format="d",
        style=style,
    )
    gain = widgets.BoundedIntText(
        value=6,
        min=-12,
        max=12,
        description="Ganho:",
        continuous_update=False,
        style=style,
    )
    center_frequency = widgets.IntSlider(
        value=11000,
        min=500,
        max=22000,
        step=500,
        description="Frequência central:",
        continuous_update=False,
        style=style,
    )
    bandwidth = widgets.IntSlider(
        value=1000,
        min=500,
        max=22000,
        step=500,
        description="Largura de banda:",
        continuous_update=False,
        style=style,
    )

    submit_button = widgets.Button(description="Atualizar", button_style="success")
    submit_button.on_click(submit)
    filter_chose_type.observe(change)
    loading_widget = widgets.Output()
    loading_gif = widgets.Image(
        value=image,
        format="gif",
        width=100,
        height=100,
    )

    display(
        filter_chose_class,
        filter_chose_type,
        file_name,
        gain,
        cutoff_frequency,
        submit_button,
    )
    display(box)
```

```python
def widgets_wah():
    # criação de caixas para deixar mais fácil a visualização das respostas
    box = box_layout()

    def submit(change):
        clear_output()
        out1.clear_output()
        out2.clear_output()
        out3.clear_output()
        with loading_widget:
            display(loading_gif)
        display(
            file_name, range_frequency, damp, wah_frequency, submit_button
        )
        display(loading_widget)
        wah_filter(
            damp.value,
            range_frequency.value[0],
            range_frequency.value[1],
            file_name.value,
            wah_frequency.value,
        )
        loading_widget.clear_output()
        display(box)

    # criação de widgets para alterar de forma interativa os filtros

    file_name = widgets.Text(
        value="piano.wav",
        description="Nome do arquivo:",
        continuous_update=False,
        style=style,
    )
    range_frequency = widgets.IntRangeSlider(
        value=[300, 5000],
        min=100,
        max=10000,
        step=100,
        description="Range de frequências:",
        disabled=False,
        continuous_update=False,
        orientation="horizontal",
        readout=True,
        readout_format="d",
        style=style,
    )
    damp = widgets.FloatSlider(
        value=0.05,
        min=0.01,
        max=0.1,
        step=0.01,
        description="Damping:",
        continuous_update=False,
        style=style,
    )
    wah_frequency = widgets.BoundedIntText(
        value=2000,
        min=100,
        max=100000,
        step=100,
        description="Frequência wah:",
        continuous_update=False,
        style=style,
    )

    submit_button = widgets.Button(description="Atualizar", button_style="success")
    submit_button.on_click(submit)

    loading_widget = widgets.Output()
    loading_gif = widgets.Image(
        value=image,
        format="gif",
        width=100,
        height=100,
    )

    display(file_name, range_frequency, damp, wah_frequency, submit_button)
    display(box)
```
